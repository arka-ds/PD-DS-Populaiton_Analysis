```python
df = df.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code'], axis =1)
```

```python
years = df.columns[1:]
total_population = df[years].sum()


plt.figure(figsize = (30 , 30))
plt.barh(years,total_population, color = 'blue')
plt.xlabel('Years', size = 20)
plt.ylabel('Total Female Population' , size = 20)
plt.title('Population Over The Time', size = 20)
plt.xticks(rotation = 90)
plt.show()
```

```python
years = df.columns[1:]
total_population = df[years].sum()


plt.figure(figsize = (30 , 30))
plt.plot(years,total_population, color = 'blue')
plt.xlabel('Years', size = 20)
plt.ylabel('Total Male Population' , size = 20)
plt.title('Population Over The Time', size = 20)
plt.xticks(rotation = 90)
plt.show()
```
